```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

```python
qc = QuantumCircuit(3,3)
```

```python
%matplotlib inline
```

```python
qc.draw(output='mpl',style='iqp')
```

```python
qc = QuantumCircuit(3,3)
qc.x(0)
qc.barrier()
qc.h(1)
qc.cx(1, 2)
qc.cx(0, 1)
qc.h(0)
qc.barrier()
qc.measure([0,1], [0,1]) # 0 to 0, 1 to 1
qc.cx(1,2)
qc.measure(2,2)
qc.draw(output='mpl',style='iqp')
```

```python
sim = Aer.get_backend('qasm_simulator')
transpiled_circuit = transpile(qc, backend=sim)
results = sim.run(transpiled_circuit).result()
plot_histogram(results.get_counts(qc))
```
